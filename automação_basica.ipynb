- Sistema de automação de tarefas simples.

Por motivos didaticos, estarei usando como exemplo uma base de dados de vendas ficticias, como se estivesse automatizando o cálculo de de vendas para enviar para a diretoria da empresa. 

obs.: para automatizar basta usar o arquivo .py no agendador de tarefas. 

```python
import pyautogui
import time
import pyperclip

pyautogui.PAUSE = 1
# Passo 1 - entrar no sistema desejado
pyautogui.hotkey('ctrl', 't')
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

# Passo 2 - navegar no sistema até o local da tarefa desejada
time.sleep(5) # Esperar a pg carregar
pyautogui.click(x=332, y=267, clicks=2)

# Passo 3 - pegar o material necessário para a tarefa
time.sleep(2) # Esperar a pg carregar
pyautogui.click(x=376, y=333)
pyautogui.click(x=1154, y=146)
pyautogui.click(x=1050, y=559)
time.sleep(3) # Esperar download
```

```python
# Passo 4 - usar a base de dados. No caso atual será calcular os dados.
import pandas as pd
tabela = pd.read_excel(r"C:\Users\pichau\Downloads\Vendas - Dez.xlsx")
faturamento = tabela['Valor Final'].sum()
quantidade = tabela['Quantidade'].sum()
display(tabela)
```

```python
# Passo 5 - enviar resultados
link = 'https://mail.google.com/'
pyautogui.hotkey('ctrl', 't')
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

# clicar no botão escrever 
time.sleep(5) # Esperar a pg carregar
pyautogui.click(x=82, y=172)

# escrever destinatário
pyautogui.write('pythonimpressionador+diretoria@gmail.com')
pyautogui.press('tab') # escolhe o e-mail
pyautogui.press('tab') # pula para o próx campo

# escrever assunto
assunto = 'Relatório de Vendas'
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

# escrever e-mail
texto = f"""
Prezados, bom dia

O faturamento foi de R$ {faturamento:,.2f};
A quantidade de produtos vendidos foi de {quantidade:,}.

Abs,
João Oliveira."""
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')


# enviar e-mail
pyautogui.hotkey('ctrl', 'enter')
```
