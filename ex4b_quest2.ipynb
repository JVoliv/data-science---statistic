<div><h2>Sobre o JSON:</h2></div>
- JSON, seu significado teórico é "Javascript Object Notation". É basicamente um formato leve de troca de informações/dados entre sistemas. É semelhante ao XML, tem a mesma utilidade, a mesma finalidade, mas é mais leve e menos detalhado. Apesar de no nome ter "Javascript", muitas linguagens suportam JSON, que é um novo método que substitui o XML. Geralmente é usado para usar solicitações AJAX para retornar dados do servidor para atualizar os dados em tempo real.

<div><h3>Vantagens do JSON:</h3></div>
    1. Leitura mais simples;
    2. Analisador(parsing) mais fácil;
    3. JSON suporta objeto;
    4. Velocidade maior na execução e transporte de dados;
    5. Arquivo com tamanho reduzido.

```python
et2019 = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/vdpb/gasolina-aviacao/vendas-gasolina-aviacao-municipio-2019.csv', sep=';')
et2018 = pd.read_csv('http://www.anp.gov.br/arquivos/dadosabertos/vc/vendas-gasolinaaviacao-municipio-2018.csv', encoding = 'ISO-8859-1', sep=';')
et2000 = pd.read_csv('http://www.anp.gov.br/arquivos/dadosabertos/vc/vendas-gasolinaaviacao-municipio-2000.csv', encoding = 'ISO-8859-1', sep =';')
```

```python
df_et = pd.concat([et2019, et2018, et2000])
df1 = df_et['ANO']
df2 = df_et['ESTADO']
df3 = df_et['Vendas']

newDF = pd.concat([df1, df2, df3], axis=1)
```

```python
df_json = newDF.to_json(orient='values')
df_json
```
