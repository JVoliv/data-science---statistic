<div><h3>(b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos
considerados? (proporção = numero de ilesos ou feridos graves/ número total de
    pessoas envolvidas no acidente)</h3><div>

Como nessa questão eu deveria apresentar as proporções por mês, pensei em usar a própria data que é apresentada no DF, mas para isso deveria primeiro transformar a coluna referente às datas para o formato de data, que pode ser visto abaixo a forma que usei isso. Tenho que admitir que tive que pesquisar um pouco pra saber como poderia fazer isso de forma eficaz e encontrei as respostas para as minhas dúvidas nos sites à seguir: https://ichi.pro/pt/trabalhando-com-data-e-hora-no-pandas-dataframe-112422831965575 e https://www.delftstack.com/pt/api/python-pandas/pandas-dataframe-dataframe.groupby-function/#:~:text=A%20função%20pandas.,o%20método%20grupo-por()%20. 

O uso do groupby foi feito para poder agrupar os dados por mês, então eu passei como parâmetro o DF referente à cada ano e suando o dt.month, para espescificar que seria por mês, claro, e retornei a soma de cada um. 

```python
df_18['data inversa'] = pd.to_datetime(df_18['data inversa'])
ilesos_18 = df_18.groupby(df_18['data inversa'].dt.month)['ilesos'].sum()
graves_18 = df_18.groupby(df_18['data inversa'].dt.month)['feridos graves'].sum()
pessoas_18 = df_18.groupby(df_18['data inversa'].dt.month)['pessoas'].sum()
```

Tendo feito isso, apenas realizei a proporção da forma como é descrito no enunciado da questão, como está abaixo com o ano de 2018. Interessante apontar que são 12 linhas, pois cada uma é referente à um mês do ano. 

Após isso, simplemente fiz o mesmo para todos os outros anos. 

```python
print("Proporção de ilesos de 2018, por mês:")
ilesos_18/pessoas_18
```

```python
print("Proporção de feridos graves de 2018, por mês:")
graves_18/pessoas_18
```

```python
df_19['data inversa'] = pd.to_datetime(df_19['data inversa'])
ilesos_19 = df_19.groupby(df_19['data inversa'].dt.month)['ilesos'].sum()
graves_19 = df_19.groupby(df_19['data inversa'].dt.month)['feridos graves'].sum()
pessoas_19 = df_19.groupby(df_19['data inversa'].dt.month)['pessoas'].sum()
```

```python
print("Proporção de ilesos de 2019, por mês:")
ilesos_19/pessoas_19
```

```python
print("Proporção de feridos graves de 2019, por mês:")
graves_19/pessoas_19
```

```python
df_20['data inversa'] = pd.to_datetime(df_20['data inversa'])
ilesos_20 = df_20.groupby(df_20['data inversa'].dt.month)['ilesos'].sum()
graves_20 = df_20.groupby(df_20['data inversa'].dt.month)['feridos graves'].sum()
pessoas_20 = df_20.groupby(df_20['data inversa'].dt.month)['pessoas'].sum()
```
