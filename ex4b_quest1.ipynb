<h1>Questão 1</h1>

- Primeiramente, devo dizer que tive bastante dificuldade de entender como poderia unir tantos dados em um só DF. O primeiro problema que enfrentei foi que as informações não "combinavam". Digo, elas não seguiam um "modelo", tendo em todos os anos, nomes diferentes pras colunas que muitas vezes tinham o mesmo tipo de dados. Por exemplo, mais abaixo vemos no DF final que há uma coluna que se chama 'NOME COORDENADOR', que no caso é o nome do pesquisador; e em alguns dos arquivos realmente estava com esse nome, mas em outros estava como esse que está agora. Dessa forma, decidi começar importando todos os arquivos que ia usar, dando preferência para os que tinham os nomes das colunas mais parecidos. Isso porque assumi que, uma vez que fosse usar o método do concat para unir todos os dados, precisaria ter o máximo de semelhança possível nos nomes das colunas. Mesmo após importar precisei realizar algumas alterações/correções nos nomes das colunas para que se encaixasse da forma que melhor atendesse à demanda. Além de corrigir os nomes das colunas, também excluí algumas colunas que não estavam fazendo nada além de ocupar espaço na tela. Apesar de não ter conseguido atender ao outro parâmetro colocado, de colocar os Anos como index, espero que algo seja considerado.

```python
# Relação de projetos de pesquisa realizados
ano_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36822', encoding='ISO-8859-1') #2019
ano_2019.rename(columns={'Campus': 'CAMPUS', 'Título': 'TÍTULO', ' Nome Coordenador':'NOME COORDENADOR'}, inplace = True)
ano_2019.drop(ano_2019.columns[[3]], axis=1,inplace=True)

ano_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33101', encoding='ISO-8859-1') #2018
ano_2018.rename(columns={'Projetos de Pesquisa Cadastrados em 2018': 'CAMPUS', 'Unnamed: 1': 'TÍTULO', 'Unnamed: 2':'NOME COORDENADOR'}, inplace = True)
ano_2018.drop(0, inplace=True)

ano_2017 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=26911', encoding='ISO-8859-1') #2017
ano_2017.rename(columns={'Projetos de Pesquisa Cadastrados em 2017': 'CAMPUS', 'Unnamed: 1': 'TÍTULO', 'Unnamed: 2':'NOME COORDENADOR'}, inplace = True)
ano_2017.drop(0, inplace=True)
```

```python
# Relação de pesquisadores
rel_Pesquisadores2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36821', encoding = "ISO-8859-1")
rel_Pesquisadores2019.rename(columns={'Coordenadores de Projetos de Pesquisa Cadastrados  por Campus': 'NOME COORDENADOR', 'Unnamed: 1':'CAMPUS'}, inplace = True)
rel_Pesquisadores2019.drop(0, inplace=True)

rel_Pesquisadores2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33103', encoding = "ISO-8859-1")
rel_Pesquisadores2018.rename(columns={'Projetos de Pesquisa Cadastrados em 2018': 'NOME COORDENADOR', 'Unnamed: 1':'CAMPUS'}, inplace = True)
rel_Pesquisadores2018.drop(0, inplace=True)

rel_Pesquisadores2017 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=26909', encoding = "ISO-8859-1")
rel_Pesquisadores2017.rename(columns={'Pesquisadores do IFNMG': 'NOME COORDENADOR', 'Unnamed: 1':'CAMPUS'}, inplace = True)
rel_Pesquisadores2017.drop(0, inplace=True)
```

```python
#Projetos de Extensão
proj_ext2019= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36482', encoding = "ISO-8859-1")
proj_ext2019.rename(columns={'Título': 'TÍTULO', 'Coordenador': 'NOME COORDENADOR', 'Campus':'CAMPUS'}, inplace = True)
proj_ext2019.drop(proj_ext2019.columns[[3]], axis=1, inplace = True)

proj_ext2018= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33131', encoding = "ISO-8859-1")

proj_ext2017= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33133', encoding = "ISO-8859-1")
```

```python
# Eventos de Extensão
ev_ext2019= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36480', encoding = "ISO-8859-1")
ev_ext2019.rename(columns={'Campus':'CAMPUS'}, inplace=True)

ev_ext2018= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33130', encoding = "ISO-8859-1")
ev_ext2018.rename(columns={' EVENTO/ SERVIÇO': 'Tipo de Evento'}, inplace=True)

ev_ext2017= pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33132', encoding = "ISO-8859-1")
ev_ext2017.drop(ev_ext2017.columns[[2]], axis=1, inplace = True)
```

```python
proj = pd.concat([ano_2019, ano_2018, ano_2017], ignore_index=True)
rel_pesq = pd.concat([rel_Pesquisadores2019, rel_Pesquisadores2018, rel_Pesquisadores2017], ignore_index=True)
proj_ext = pd.concat([proj_ext2019, proj_ext2018, proj_ext2017], ignore_index=True)
ev_ext = pd.concat([ev_ext2019, ev_ext2018, ev_ext2017], ignore_index=True)
```

```python
concatF = pd.concat([proj, rel_pesq, proj_ext, ev_ext], ignore_index=True)
newDF = concatF.set_index(['CAMPUS'])
newDF
```
